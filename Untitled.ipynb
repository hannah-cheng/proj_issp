```python
# import packages
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

low, high = 0, 100  # range of data points
n = 100  # number of data points

f = lambda: np.random.randint(low=low, high=high, size=n)
x, y = f(), f()  # generate some random data points

model = LinearRegression()\
            .fit(x.reshape((-1, 1)), y.reshape((-1, 1)))  # find the target
m, b = model.coef_, model.intercept_  # get the slope and bias of line

# draw picture
line_x, line_y = [low, high], [m * low + b, m * high + b]
plt.scatter(x, y, alpha=0.5)
plt.plot(line_x, line_x, color='red', linestyle='dashed')
plt.show()
```
